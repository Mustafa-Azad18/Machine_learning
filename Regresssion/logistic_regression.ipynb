```python
import pandas as pd 
import numpy as np 
import  matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
from sklearn.datasets import make_classification 
```

```python
##crate the dataset 
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=3)
```

```python
X
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
print(y_pred)
```

```python
lr.predict_log_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
print("y_test:", y_test[:10])   # first 10 true labels
print("y_pred:", y_pred[:10])   # first 10 predicted labels
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

 ### HYPERPARAMETER TUNING AND CROSS VALIDATION

```python
model= LogisticRegression()
penalty= ['l1','l2','elasticnet']
c_vlaues=[100,10,1.0,0.1,0.01]
solver=[ 'newton_cg', 'lbfgs','newton_cholesky', 'sag', 'saga']
```

```python
params= dict(penalty=penalty,C=c_vlaues,solver=solver)
```

```python
params
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
#Grid_search
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

### Randomized SearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model= LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5)
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

# Logistic Regression for multiclass classification

```python
##crate the dataset 
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=3)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
Lr=LogisticRegression(multi_class='ovr')
Lr.fit(X_train,y_train)
y_pred=Lr.predict(y_test)
```

```python
Lr=LogisticRegression(multi_class='ovr')
```

```python
Lr.fit(X_train,y_train)
y_pred=Lr.predict(X_test)
```

```python
print("y_test:", y_test[:10])   # first 10 true labels
print("y_pred:", y_pred[:10])   # first 10 predicted labels
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
