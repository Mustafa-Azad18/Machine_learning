```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
X= np.sort(np.random.rand(20, 1)* 10, axis=0)
y= np.sin(X).ravel()+ np.random.randn(20) *0.1
```

```python
# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_lin_pred = lin_reg.predict(X)
```

```python
# Polynomial Regression (Degree 15)
poly_features = PolynomialFeatures(degree=15)
X_poly = poly_features.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
y_poly_pred = poly_reg.predict(X_poly)
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
plt.scatter(X, y, color='black', label='Data')
plt.plot(X, y_lin_pred, color='blue', label='Linear fit')
plt.title('Linear Regression')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X, y, color='black', label='Data')
plt.plot(X, y_poly_pred, color='red', label='Polynomial fit (degree 15)')
plt.title('Polynomial Regression')
plt.legend()
```
